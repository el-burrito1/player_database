```python
import urllib2
import pandas as pd
import re
import os
import numpy as np

opener = urllib2.build_opener()
opener.addheaders = [('User-Agent', 'Mozilla/5.0')]
teams = 'http://www.espn.com/college-football/teams'
page = opener.open(teams)

from bs4 import BeautifulSoup
teams = BeautifulSoup(page , "html5lib")

div = teams.select("div.span-2.last")
div_zero = div[0]
conf_divs = div_zero.find_all('div' , class_='mod-container mod-open-list mod-teams-list-medium mod-no-footer')

bigsky_div = conf_divs[0]
bigsouth_div = conf_divs[1]
caa_div = conf_divs[2]
ivy_div = conf_divs[3]
meac_div = conf_divs[4]
missourivalley_div = conf_divs[5]
northeast_div = conf_divs[6]
ohiovalley_div = conf_divs[7]
patriot_div = conf_divs[8]
pioneer_div = conf_divs[9]
southern_div = conf_divs[10]
southland_div = conf_divs[11]
swac_div = conf_divs[12]

all_conf = [bigsky_div,
              bigsouth_div,
              caa_div,
              ivy_div,
              meac_div,
              missourivalley_div,
              northeast_div,
              ohiovalley_div,
              patriot_div,
              pioneer_div,
              southern_div,
              southland_div,
              swac_div]
```

```python
for conf in all_conf:
    filepath = conf.find_all('h4')[0].string
    os.makedirs(filepath)
        
    _name = []
    _teams = []
    _rosters = []
    
    for i in conf.find_all('a' , class_='bi'):
        _name.append(i)
    for i in _name:
        _teams.append(i.contents[0])
    
    for i in conf.find_all('a' , href = re.compile('roster')):
        _rosters.append(i['href'])
        
    for team, y in zip(_teams,_rosters):
        roster = 'http://www.espn.com' + y
        page = opener.open(roster)
        soup = BeautifulSoup(page)
        players = soup.find_all('tr' , class_=['oddrow','evenrow'])

        A = []
        B = []
        C = []
        D = []
        E = []
        F = []
        G = []
        H = []
    
        for player in players:
            stats = player.find_all('td')
            A.append(stats[1].find(text=True))
            B.append(stats[0].find(text=True))
            C.append(stats[2].find(text=True))
            D.append(stats[3].find(text=True))
            E.append(stats[4].find(text=True))
            F.append(stats[5].find(text=True))
            G.append(stats[6].find(text=True))
    
        D = [item.replace('-','') for item in D]
        D = [item + '0' if len(item)==3 else item[0:1] + '0' + item[1:] + '0' for item in D]
        D = [int(item) for item in D]
    
        df = pd.DataFrame(A,columns=['Name'])
        df['Number'] = B
        df['Position'] = C
        df['Height'] = D
        df['Weight'] = E
        df['Class'] = F
        df['Hometown'] = G

        df.to_csv(filepath + '/' + team + '.csv',index=False,encoding='utf-8')
        
```
